# Estimación del área del espejo de agua

Este notebook documenta la metodología utilizada para estimar el área del espejo de agua de la Laguna de Fúquene a partir de imágenes satelitales, así como las decisiones técnicas adoptadas durante el análisis.

## 1. Objetivo del análisis

El objetivo principal es estimar la variación temporal del área cubierta por agua dentro del Área de Interés (AOI) definida, utilizando datos de Observación de la Tierra.

Este análisis busca identificar tendencias y patrones generales en el comportamiento del espejo de agua, más que reproducir cifras oficiales exactas.

## 2. Observación de la Tierra (EO) como fuente de datos

La Observación de la Tierra (Earth Observation, EO) permite analizar fenómenos ambientales de forma sistemática y consistente en el tiempo.

Sin embargo, los datos EO presentan limitaciones inherentes:
- Dependencia de condiciones atmosféricas
- Resolución espacial finita
- Sensibilidad a decisiones metodológicas

Este proyecto asume estas limitaciones como parte del análisis.

## 3. Selección del conjunto de datos satelitales

Se seleccionó la misión Landsat como fuente principal de datos debido a:
- Su extensa cobertura temporal (décadas)
- La consistencia entre sensores
- Su uso frecuente en estudios hidrológicos

Aunque Sentinel-2 ofrece mayor resolución espacial, su cobertura temporal más corta limita su utilidad para análisis históricos de largo plazo.

## 4. Identificación del agua mediante información espectral

El agua fue identificada utilizando un índice espectral calculado a partir de las bandas verde y del infrarrojo cercano (NIR) de Landsat.

Este tipo de índice se basa en el contraste espectral entre:
- Alta reflectancia del agua en la banda verde
- Baja reflectancia del agua en el infrarrojo cercano

La clasificación resultante es una aproximación binaria del estado agua / no agua.

## 5. Umbral y clasificación

Se utilizó un umbral fijo para clasificar los píxeles como agua.

Esta decisión implica un compromiso:
- Umbrales más bajos incluyen áreas con vegetación o humedad
- Umbrales más altos excluyen zonas someras o turbias

Se priorizó la consistencia temporal sobre la optimización local del umbral.

## 6. Cálculo del área

El área del espejo de agua se estimó contando los píxeles clasificados como agua dentro del AOI y multiplicando por el área espacial de cada píxel.

Este enfoque implica que:
- El área se cuantiza a la resolución del sensor (30 m)
- Pequeñas variaciones espaciales no son capturadas

El resultado representa una estimación discreta del área.

## 7. Resolución temporal y datos faltantes

Inicialmente se exploró una resolución mensual.

Este enfoque permitió identificar:
- Meses sin observaciones válidas
- Impacto significativo de la nubosidad
- Incremento en la variabilidad y en los datos faltantes

Estas observaciones influyeron en la interpretación de la serie temporal.

## 8. Interpretación de datos faltantes y valores atípicos

Los datos faltantes y valores extremos no fueron interpolados ni corregidos artificialmente.

En este proyecto se interpretan como:
- Consecuencia de nubosidad persistente
- Limitaciones del sensor
- Resultados de la metodología de clasificación

Estos valores aportan información sobre la calidad y disponibilidad del dato.

## 9. Comparación con cifras oficiales

Las estimaciones obtenidas son significativamente menores que algunas cifras oficiales reportadas para la laguna.

Esta discrepancia puede explicarse por:
- Diferencias en la definición conceptual del área
- Inclusión o exclusión de zonas inundables
- Diferencias metodológicas entre mediciones in situ y EO

La comparación se utiliza como referencia, no como validación estricta.

## 10. Aprendizajes clave

Este análisis permitió identificar aprendizajes relevantes:

- Las decisiones espaciales tienen un impacto mayor que las decisiones temporales
- Mayor resolución no garantiza mejores resultados históricos
- Los datos EO requieren interpretación contextual

El valor del análisis reside tanto en el resultado como en el proceso de toma de decisiones.

## 11. Conclusiones

La estimación del área del espejo de agua mediante EO es una aproximación útil para analizar tendencias de largo plazo, pero no sustituye mediciones directas ni definiciones administrativas.

Este proyecto demuestra la capacidad de:
- Trabajar con datos incompletos
- Evaluar múltiples enfoques
- Tomar decisiones metodológicas informadas
- Comunicar resultados con sus limitaciones
