# Definición del Área de Interés (AOI)

Este notebook documenta el proceso de definición, validación y revisión del Área de Interés (AOI) utilizada para el análisis del espejo de agua de la Laguna de Fúquene.

## 1. Importancia del Área de Interés

En análisis de Observación de la Tierra, la definición del Área de Interés (AOI) es una de las decisiones metodológicas más críticas.

Un AOI mal definido puede generar:
- Subestimaciones o sobreestimaciones sistemáticas del área
- Series temporales inconsistentes
- Resultados que no son comparables con fuentes oficiales

Por esta razón, este proyecto documenta explícitamente cómo y por qué se define el AOI.

## 2. Área de estudio

El área de estudio corresponde a la Laguna de Fúquene, ubicada en el altiplano cundiboyacense, Colombia.

La laguna presenta una morfología compleja, con variaciones históricas en su extensión, zonas inundables y sectores con vegetación emergente, lo que dificulta una delimitación precisa basada únicamente en una imagen puntual.

## 3. Definición inicial del polígono

La definición inicial del polígono se realizó a partir de referencias cartográficas generales y visualización de imágenes satelitales recientes.

Durante la exploración inicial se identificó que:
- El polígono cubría únicamente una fracción de la laguna
- Sectores históricamente inundables quedaban excluidos
- El área calculada era sustancialmente menor a valores reportados por fuentes oficiales

## 4. Revisión y corrección del AOI

A partir de estas discrepancias, se decidió revisar y corregir el AOI.

La corrección incluyó:
- Expansión del polígono para cubrir la extensión completa conocida de la laguna
- Inclusión de zonas inundables históricas
- Validación visual utilizando imágenes multitemporales

Este proceso redujo una fuente importante de sesgo sistemático en los cálculos de área.

## 5. Uso de un AOI fijo en el tiempo

El análisis utiliza un AOI fijo para toda la serie temporal.

Esta decisión se tomó para:
- Garantizar comparabilidad entre distintos años
- Evitar que cambios en el polígono introduzcan variabilidad artificial
- Asegurar que las diferencias observadas correspondan a cambios en el espejo de agua y no a modificaciones del área analizada

Se priorizó reproducibilidad y consistencia por sobre una representación dinámica del contorno.

## 6. Escala y resolución espacial

El AOI fue definido considerando la resolución espacial de los datos Landsat (30 metros).

A esta escala:
- Los bordes del cuerpo de agua son necesariamente aproximados
- Pequeñas variaciones locales no pueden ser capturadas con precisión

El AOI no pretende representar límites administrativos o legales, sino un marco espacial consistente para análisis multitemporal.

## 7. Limitaciones conocidas

La definición del AOI presenta limitaciones que se reconocen explícitamente:

- No captura expansiones extremas fuera del polígono definido
- Incluye áreas que pueden no estar permanentemente inundadas
- Depende parcialmente de interpretación visual

Estas limitaciones son consideradas en la interpretación de los resultados.

## 8. Aprendizajes metodológicos

La definición del AOI fue uno de los principales puntos de aprendizaje del proyecto.

Este proceso permitió comprender que:
- Errores espaciales iniciales se amplifican en análisis temporales
- La validación visual es indispensable incluso con datos satelitales
- Las discrepancias con cifras oficiales no siempre indican errores de cálculo, sino diferencias conceptuales en la definición del área
